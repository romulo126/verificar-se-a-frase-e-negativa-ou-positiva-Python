```python
def adicionarnovasstopwords(arraystopword):
    for(palavra) in arraystopword:
        stopwords.append(palavra)
```

```python
stopwords=nltk.corpus.stopwords.words('portuguese')
adicionarnovasstopwords(['tão','vou','vai','com'])
```

```python
def removestopword(text):
    frases=[]
    for(palavras,emocao) in text:
        semstop=[p for p in palavras.split() if p not in stopwords]
        frases.append((semstop, emocao))
    return frases
```

```python
def aplicarstemmer(text,modo=0):
    stemmer=nltk.stem.RSLPStemmer()
    frasessteme=[]
    if(modo == 0):
        for(palavas,emocao) in text:
            comsteamming = [str(stemmer.stem(p)) for p in palavas.split() if p not in stopwords]
            frasessteme.append((comsteamming,emocao))
        
    else:
        for(palavras) in text.split():
            comsteamming = [p for p in palavras.split()]
            frasessteme.append(str(stemmer.stem(comsteamming[0])))
    return frasessteme        
```

```python
def busapalavras(frases):
    todaspalavras=[]
    for (palavras,emocao) in frases:
        todaspalavras.extend(palavras)
    return todaspalavras
```

```python
def buscafrequencia(palavras):
    palavras = nltk.FreqDist(palavras)
    return palavras
```

```python
def extratordepalavras(documento):
    doc = set(documento)
    caracteristicas={}
    for palavras in palavrasunicas:
        caracteristicas['%s' % palavras] = (palavras in doc)
    return caracteristicas
```

```python
def returnporcentagem(text):
    porcentage =[]
    text = clasificador.prob_classify(text)
    for classe in text.samples():
        porcentage.append(text.prob(classe))
    return porcentage
```

```python
def retornaposicaodamaioremocao(array):
    chave=0
    for(position) in array:
        if not 'chavetemp' in locals():
            chavetemp=chave
            temp=position
        else:
            if(temp<position):
                chavetemp=chave
                temp=position
        chave +=1
    return [chavetemp,temp]
```
